```python
import pandas as pd
```

```python
df= pd.read_csv("customer_shopping_behavior.csv")
df.head(20)
```

```python
df.info()
```

```python
df.describe(include="all") #gives you summary statistics of all columns (num+categories)
```

```python
df.isnull().sum() #tells you many null values there are (37 null values for review rating)
```

```python
df["Review Rating"]= df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median())) #lambda is a way to define a small function
#For each category, replace missing review ratings with the median rating of that category (clothing and footwear)
#x stands for review rating for the category
#takes median based on category, not the same review for both footwear and clothing
```

```python
df.isnull().sum() #no more null values left
```

```python
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(" ", "_")
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})#makes it easier to see the data, columns gives u the name
df.columns
```

```python
#creating a new column called age_group
labels= ["young adult", "adult", "middle aged", "elderly"]
df["age_group"]= pd.qcut(df["age"], q=4, labels=labels)
df[["age", "age_group"]].head(15)
df
```

```python
#another feature
frequency= { #dictionary that maps word to num days, the map function replaces text with corresponding number of days
    "Fortnightly": 14,
    "Weekly": 7,
    "Monthly": 30,
    "Quarterly": 90,
    "Bi-weekly": 14,
    "Annually": 365,
    "Every 3 Months": 90
}
df["purchase_frequency_days"]= df["frequency_of_purchases"].map(frequency)
```
